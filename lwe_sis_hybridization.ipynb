```python
import numpy as np
```

Some parameters

```python
n = 64
m = 128
q = 10007
alpha = 0.005

M = np.random.randint(0, q, (n, m))
```

```python
def discrete_gaussian(size):
    return np.round(np.random.normal(0, alpha * q, size)).astype(int) % q
```

```python
def signal_function(y):
    """Hint algorithm S(y) to generate a signal[cite: 273, 280]."""
    # Simplified version of the signal function logic
    return (y > (q // 4)) and (y < (3 * q // 4))
```

```python
def robust_extractor(x, sigma):
    """Extracts the shared bit[cite: 281]."""
    return ((x + sigma * ((q - 1) // 2)) % q) % 2
```

Alice's step: LWE

```python
s_A = np.random.randint(0, q, (1, n))
e_A = discrete_gaussian((1, m))
P_A = (s_A @ M + 2 * e_A) % q
```

Bob step: SIS

```python
s_B = discrete_gaussian((m, 1))
P_B = (M @ s_B) % q
```

Computing the common secret

```python
# Bob
K_B = (P_A @ s_B) % q

# Alice
K_A = (s_A @ P_B) % q
sigma = signal_function(K_A[0,0])
```

Key extraction

```python
root_key_alice = robust_extractor(K_A[0,0], sigma)
root_key_bob = robust_extractor(K_B[0,0], sigma)

print(root_key_alice)
print(root_key_bob)

assert root_key_alice == root_key_bob
```
