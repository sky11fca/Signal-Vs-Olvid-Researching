```python
import random
```

# Global Parameters

```python
n = 256
q = 7681
```

```python
def dbl(x):
    e_bar = random.choices([-1, 0, 1], weights = [0.25, 0.5, 0.25])[0]
    return (2*x - e_bar) % (2*q)
```

```python
def cross_rounding(v_bar):
    return(int((4 / (2*q)) * v_bar)) % 2
```

```python
def reconcile(v_prime, c):
    if(c==0):
        return 1 if (q/2 <= (2* v_prime) % (2*q) < 3*q/2) else 0
    else:
        return 1 if(0 <= (2*v_prime) % (2*q) < q/2) else 0
```

# Protocol simulation

Bob generates a secret value $v$ and reconciliation bits $c$

```python
v_bob = random.randint(0, q-1)
v_bar = dbl(v_bob)
c_hint = cross_rounding(v_bar)
root_key_bob = (v_bar // q)%2
```

Computing the Alice root key

```python
e_A = random.randint(-10, 10)
v_prime = (v_bob + e_A) % q

root_key_alice = reconcile(v_prime, c_hint)
```

```python
print(root_key_alice)
print(root_key_bob)

assert root_key_bob == root_key_alice
```
